```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import tqdm
import pandas as pd
import json

ua = UserAgent()
session = requests.session()
```

## Женский стендап

```python
all_jokes = []

for num in range(0, 9):
  url = f'https://citaty.info/tv/zhenskii-stendap?page={num}'
  req = session.get(url, headers={'User-Agent': ua.random})
  page = req.text
  soup = BeautifulSoup(page, 'html.parser')
  jokes = soup.find_all('div', {'class': 'node__content'})
  for j in jokes:
    short_text = j.find('div', {'class': 'field-item even last'}).text
    short_text = short_text.replace('\xa0', " ")
    all_jokes.append(short_text)
```

```python
good_women = []

for i in all_jokes:
  if len(i) < 300:
    good_women.append(i)

len(good_women)
```

## Основной стендап

```python
all_jokes = []

for num in tqdm.tqdm(range(0, 36)):
  url = f'https://citaty.info/tv/stand-up?page={num}'
  req = session.get(url, headers={'User-Agent': ua.random})
  page = req.text
  soup = BeautifulSoup(page, 'html.parser')
  jokes = soup.find_all('div', {'class': 'node__content'})
  for j in jokes:
    short_text = j.find('div', {'class': 'field-item even last'}).text
    short_text = short_text.replace('\xa0', " ")
    all_jokes.append(short_text)
```

```python
good_gen = []

for i in all_jokes:
  if len(i) < 300:
    good_gen.append(i)

len(good_gen)
```

## BBF

```python
all_jokes = []

for num in tqdm.tqdm(range(1, 26)):
  url = f'https://bbf.ru/quotes/source/70005/page/{num}/'
  req = session.get(url, headers={'User-Agent': ua.random})
  page = req.text
  soup = BeautifulSoup(page, 'html.parser')
  jokes = soup.find_all('div', {'class': 'sentence__body'})
  for j in jokes:
    # short_text = j.find('div', {'class': 'field-item even last'}).text
    # short_text = short_text.replace('\xa0', " ")
    all_jokes.append(j.text)
```

```python
good_bbf = []

for i in all_jokes:
  if len(i) < 300:
    good_bbf.append(i)

len(good_bbf)
```

```python
df_up = pd.DataFrame(good_women + good_gen + good_bbf + good_bbf2)
df_up
```

## Пездуза

```python
with open('result.json', 'r', encoding='utf-8') as f:
  data = json.load(f)
```

```python
len(data['messages'])
```

```python
data['messages'][0]
```

```python
texts = []

for message in data['messages']:
  if 'photo' in message.keys():
    continue
  else:
    texts.append(message['text'])
    for i in message['text']:
      if type(i) == dict:
        texts.append(i['text'])
```

```python
jokes = []

for i in texts:
  if type(i) != list and len(i) < 300:
    jokes.append(i)
```

```python
len(jokes)
```

```python
pezd = pd.DataFrame(jokes)
pezd
```

```python
pezd.to_csv('pezduza.csv', index=False)
```

## Объединение юмористичсеких аугментаций

```python
standup = pd.read_csv('df_up.csv', sep='\t')
pezduza = pd.read_csv('pezduza.csv', sep='\t')
ache = pd.read_csv('ache.csv', sep='\t')
```

```python
all_jokes_aug = pd.concat([standup, ache, pezduza])
all_jokes_aug.columns = ['text']
all_jokes_aug['is_joke'] = 1
all_jokes_aug
```

## Аугментация неюмористических текстов

```python
df_0_without_used = pd.read_csv('df_0_without_used.csv', sep='\t')
df_0_without_used
```

```python
aug_0 = df_0_without_used.sample(n=1874)
```

```python
aug_0_ = aug_0.text.values
len(list(aug_0_))
```

```python
train = pd.read_csv('train_30k.csv', sep='\t')
tr_0 = train.text.values
```

```python
from collections import Counter

common_items = list((Counter(aug_0_) & Counter(tr_0)).elements())
common_items
```

```python
for t in common_items:
  aug_0 = aug_0[aug_0.text != t]

aug_0
```

## Объединение шуток и нешуток

```python
augmentations = pd.concat([aug_0, all_jokes_aug])
augmentations
```

```python
del augmentations['id']
```

```python
augmentations.to_csv('augmentations.csv', sep='\t', index=False)
```
